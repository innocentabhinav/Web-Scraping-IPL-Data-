```python

#matchSummary_df["Winner"] = matchSummary_df['result'].str.split().str[0].

def extract_first_two_words(text):
  return text.split(' ')[0] + ' ' + text.split(' ')[1]

matchSummary_df['Winner'] = matchSummary_df['result'].apply(extract_first_two_words)
```

```python
matchSummary_df["Winner"].unique()
```

```python
matchSummary_df.sample(9)
```

```python
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Super Kings', 'Chennai Super Kings')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Mumbai won', 'Mumbai Indians')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Royals won', 'Rajasthan Royals')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Titans won', 'Gujrat Titans')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Punjab Kings', 'Punjab Kings')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Super Giants', 'Lucknow Super Giants')

matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Capitals won', 'Delhi Capitals')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('Sunrisers won', 'Sunrisers Hyderabad')

matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('RCB won', 'Royal Challengers Banglore')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('KKR won', 'Kolkata Knight Riders')
matchSummary_df['Winner'] = matchSummary_df['Winner'].replace('No', 'No result')
```
